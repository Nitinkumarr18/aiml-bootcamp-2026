```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Define student data
students = {
    "name": ["Aarav", "Riya", "Kabir", "Meera", "Dev", "Sara"],
    "study_hours": [6, 3, 8, 2, 5, 7],
    "attendance": [85, 70, 90, 60, 78, 92],
    "previous_marks": [78, 55, 88, 35, 65, 84],
    "final_marks": [82, 58, 91, 40, 70, 88],
}


```

```python
# Convert dictionary to DataFrame (so we can use iterrows)
df = pd.DataFrame(students)

# Scatter plot: Study Hours vs Previous Marks
plt.figure(figsize=(7, 5))
plt.scatter(df["study_hours"], df["previous_marks"], color="blue")

# Label each point with student name
for i, row in df.iterrows():
    plt.text(row["study_hours"] + 0.1, row["previous_marks"] + 0.5, row["name"])

plt.xlabel("Study Hours")
plt.ylabel("Previous Marks")
plt.title("Students as Data Points")
plt.grid(True)
plt.show()
```
